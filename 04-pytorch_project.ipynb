### 1. Package and Module Installation

First, let's pool all package and module that you'll need in the installation section below.

```python
# Write any package/module installation that you need
# pip install goes here, this helps declutter your output below
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn import datasets
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader, TensorDataset
```

### 2. Data Loading and Preprocessing

Load the chosen dataset and preprocess it for deep learning.

```python
# Write your preprocessing here
# Load your data
digits = datasets.load_digits()

# Flatten the images
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))

# split dataset to training, test (70,30)
x_train, x_test, y_train, y_test = train_test_split(data, digits['target'], test_size=0.2, random_state=42)

# Convert to PyTorch tensors
x_train_tensor = torch.tensor(x_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)
x_test_tensor = torch.tensor(x_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)

# create pytorch dataset from scikit-learn load-digits dataset
train_digits = TensorDataset(x_train_tensor, y_train_tensor)
test_digits = TensorDataset(x_test_tensor, y_test_tensor)

# using dataloader to make a batch
train_loader = DataLoader(dataset=train_digits, batch_size=64, shuffle=True)
test_loader = DataLoader(dataset=test_digits, batch_size=64, shuffle=True)
```

### 3. Model Building

Define your deep learning model's architecture.

```python
# Write your code here for Model Building here

# define your Deep Learning Model here, training is in the next section.
class LoadDigitModel(nn.Module):
  def __init__(self, input_size, hidden_size, output_size):
      super(LoadDigitModel, self).__init__()
      self.layer1 = nn.Linear(input_size, hidden_size)
      self.relu = nn.ReLU()
      self.layer2 = nn.Linear(hidden_size, output_size)

  def forward(self, x):
        # Flatten the image tensors using reshape
        x = self.layer1(x)
        x = self.relu(x) # apply relu activation function
        x = self.layer2(x)
        return x
```

### 4. Model Training

Train your model and evaluate its performance using validation data.

```python
# Write your code here for Model Training here

# Instantiate the model, loss function, and optimizer
input_size = x_train.shape[1] # number of features in training data as the input size for the neural network
hidden_size = 64 # number of neurons in the hidden layer
output_size = 10 # from 0 to 9 classification

model = LoadDigitModel(input_size, hidden_size, output_size)
criterion = nn.CrossEntropyLoss() # lost function for classification task
optimizer = optim.Adam(model.parameters(), lr=0.001) #optimizer to update weight and bias

# Define accuracy function
def accuracy(outputs, labels):
    _, preds = torch.max(outputs, dim=1)
    return torch.tensor(torch.sum(preds == labels).item() / len(preds))

#create the training loop
epochs = 10

for epoch in range(epochs):
    model.train()
    for images, labels in train_loader:
        # Generate predictions
        output = model(images)
        # calculate the loss value
        loss = criterion(output, labels)
        # gradient descent and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, 10, loss.item()))
```

### 5. Model Evaluation
Evaluate your model's performance on the test data using the grading scheme defined above.

```python
# Write your code here for Model Evaluation here


#define the iteration

#create the training loop
with torch.no_grad():
    accum_acc = 0
    for images, labels in test_loader:
        outputs = model(images)
        loss = criterion(outputs, labels)
        acc = accuracy(outputs, labels)
        accum_acc += acc

    print('Test loss: {:.4f}, Test accuracy: {:.4f}'.format(loss.item(), accum_acc/len(test_loader)))
```
